## 2 Tidying up the Score_card

Here we pick UGDS as a criteria of judging big city, sum of different races(by adding up UGD races through excel) to judge diversity, and C150_M(max of C150_4 and C150_L4) to judge quality of colleges. 

```python
Score_Selected = Score_card[['INSTNM','CITY','UGDS','Diversity','ADM_RATE','C150_M']]
Score_Selected
```

We delete the NaN values for nonsense.

```python
Score_Selected = Score_Selected.dropna(axis =0, how = 'any')
```

First we adjust the columns' name of crime to merge it with selected score.

```python
CrimeRating.rename(columns={'City':'CITY'},inplace = True)
CrimeRating = CrimeRating.drop(columns='State')
CrimeRating
```

```python
Merge = pd.merge(CrimeRating,Score_Selected ,how = 'inner')
Merge
```

```python
UGDS = Merge.sort_values(by=['UGDS'], ascending=False, ignore_index = True)
Diversity = Merge.sort_values(by=['Diversity'], ascending=False, ignore_index = True)
ADM = Merge.sort_values(by=['ADM_RATE'],ascending=False,  ignore_index = True)
C150 = Merge.sort_values(by=['C150_M'],ascending=False, ignore_index = True)
```
